**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
movie_revenue = movie_data.groupby('release_year')['revenue'].sum()
movie_revenue
```

```python
movie_budget=movie_data.groupby('release_year')['budget'].sum()
movie_budget
```

```python
movie_gain = movie_revenue-movie_budget
movie_gain.plot()
movie_budget.plot()
movie_revenue.plot()
```
